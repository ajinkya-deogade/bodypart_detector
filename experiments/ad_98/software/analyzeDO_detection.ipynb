```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook

def simpleaxis(ax):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
```

```python
coordinate_header = ['FrameNumber', 'MouthHook_x', 'MouthHook_y', 'LeftMHhook_x', 'LeftMHhook_y',
                     'RightMHhook_x', 'RightMHhook_y', 'LeftDorsalOrgan_x', 'LeftDorsalOrgan_y',
                     'RightDorsalOrgan_x', 'RightDorsalOrgan_y', 'CenterBolwigOrgan_x', 'CenterBolwigOrgan_y',
                     'LeftBolwigOrgan_x', 'LeftBolwigOrgan_y', 'RightBolwigOrgan_x', 'RightBolwigOrgan_y']
meta_data_header = ['FrameNumber', 'Time', 'Stage_x', 'Stage_y', 'Centroid_x', 'Centroid_y',
                    'Midpoint_x', 'Midpoint_y', 'Head_x', 'Head_y', 'Tail_x', 'Tail_y','MouthHook_x', 'MouthHook_y',
                    'LeftMHhook_x', 'LeftMHhook_y','RightMHhook_x', 'RightMHhook_y', 
                    'LeftDorsalOrgan_x', 'LeftDorsalOrgan_y', 'RightDorsalOrgan_x', 'RightDorsalOrgan_y',
                    'CenterBolwigOrgan_x', 'CenterBolwigOrgan_y','LeftBolwigOrgan_x', 'LeftBolwigOrgan_y',
                    'RightBolwigOrgan_x', 'RightBolwigOrgan_y', 'V9_x', 'V9_y', 'V10_x', 'V10_y', 'V11_x','V11_y',
                    'V12_x', 'V12_y', 'V13_x', 'V13_y', 'V14_x', 'V14_y', 'V15_x', 'V15_y', 'V16_x', 'V16_y',
                    'MouthHook_votes', 'LeftMHhook_votes', 'RightMHhook_votes', 'LeftDorsalOrgan_votes',
                    'RightDorsalOrgan_votes', 'CenterBolwigOrgan_votes', 'LeftBolwigOrgan_votes', 'RightBolwigOrgan_votes',
                    'V9_votes', 'V10_votes', 'V11_votes', 'V12_votes', 'V13_votes', 'V14_votes', 'V15_votes',
                    'V16_votes', 'Num_Key_points']

path_all = '../expts/dataCollectedOn_20180417/dataCollectedOn_2018041_grp_1/20180529_150721/'
path_1 = 'Rawdata_20180417_083618_20180420_104633'
coordinates_1 = pd.read_csv(os.path.join(path_all,path_1,'Rawdata_20180417_083618_20180420_104633_Coordinates.csv'),
                        sep=',', names=coordinate_header)
coordinates_1 = coordinates_1.round(0)
start_frame_1 = coordinates_1.loc[0,'FrameNumber'] - 1
coordinates_1.set_index('FrameNumber', inplace=True)
meta_data_1 = pd.read_csv(os.path.join(path_all,path_1,'Metadata_20180530_103813.txt'),
                          sep=',', header=0, names=meta_data_header)
meta_data_1.loc[:,'FrameNumber'] = meta_data_1.loc[:,'FrameNumber'] + start_frame_1
meta_data_1.set_index('FrameNumber', inplace=True)
meta_data_1_1 = meta_data_1.loc[coordinates_1.index, coordinates_1.columns.values]

path_2 = 'Rawdata_20180417_084502_20180423_075225'
coordinates_2 = pd.read_csv(os.path.join(path_all,path_2,'Rawdata_20180417_084502_20180423_075225_Coordinates.csv'),
                        sep=',', names=coordinate_header)
coordinates_2 = coordinates_2.round(0)
start_frame_2 = coordinates_2.loc[0,'FrameNumber'] - 1
coordinates_2.set_index('FrameNumber', inplace=True)
meta_data_2 = pd.read_csv(os.path.join(path_all,path_2,'Metadata_20180530_084923.txt'),
                          sep=',', header=0, names=meta_data_header)
meta_data_2.loc[:,'FrameNumber'] = meta_data_2.loc[:,'FrameNumber'] + start_frame_2
meta_data_2.set_index('FrameNumber', inplace=True)
meta_data_2_1 = meta_data_2.loc[coordinates_2.index, coordinates_2.columns.values]
```

```python
distance_header = ['MouthHook', 'LeftMHhook',
                   'RightMHhook', 'LeftDorsalOrgan', 'RightDorsalOrgan',
                   'CenterBolwigOrgan','LeftBolwigOrgan', 'RightBolwigOrgan']

temp_1 = (meta_data_1_1 - coordinates_1).apply(lambda x: x**2)
distance_1 = pd.DataFrame([], columns=distance_header, index=coordinates_1.index.values)
for head in distance_header:
    not_present = np.array([])
    distance_1.loc[:, head] = (temp_1.loc[:,head+'_x'] + temp_1.loc[:,head+'_y']).apply(lambda x: np.sqrt(x))
    not_present = coordinates_1.loc[:, head+'_x'] == -1
    if np.where(not_present)[0].size != 0:
        distance_1.loc[not_present, head] = np.NaN
# distance_1['FrameNumber'] = coordinates_1['FrameNumber']

temp_2 = (meta_data_2_1 - coordinates_2).apply(lambda x: x**2)
distance_2 = pd.DataFrame([], columns=distance_header, index=coordinates_2.index.values)
for head in distance_header:
    not_present = np.array([])
    distance_2.loc[:, head] = (temp_2.loc[:,head+'_x'] + temp_2.loc[:,head+'_y']).apply(lambda x: np.sqrt(x))
    not_present = coordinates_2.loc[:, head+'_x'] == -1
    if np.where(not_present)[0].size != 0:
        distance_2.loc[not_present, head] = np.NaN
# distance_2['FrameNumber'] = coordinates_2['FrameNumber']
```

```python
# distance_1.boxplot(figsize=(10, 10), column=distance_header, rot=45)
distance_1_melt = pd.melt(distance_1, id_vars=['FrameNumber'], value_vars=distance_header)
fig = plt.figure(figsize=(8,5), dpi=150)
ax = sns.boxplot(x='variable', y='value', data=distance_1_melt, fliersize=0)
# ax = sns.violinplot(x='variable', y='value', data=distance_1_melt, cut=0.25, inner='box', scale='count')
ax = sns.swarmplot(x='variable', y='value', data=distance_1_melt, color=".4", size=4)
simpleaxis(ax)
plt.xticks(rotation=30);
plt.xlabel('');
plt.ylabel('Distance (pixels)');
plt.ylim([0, 500]);
```

```python
# distance_2.boxplot(figsize=(10, 10), column=distance_header, rot=45)
distance_2_melt = pd.melt(distance_2, id_vars=['FrameNumber'], value_vars=distance_header)
fig = plt.figure(figsize=(8,5), dpi=150)
ax = sns.boxplot(x='variable', y='value', data=distance_2_melt, fliersize=0)
# ax = sns.violinplot(x='variable', y='value', data=distance_2_melt, cut=0.25, inner='box', scale='count')
ax = sns.swarmplot(x='variable', y='value', data=distance_2_melt, color=".4", size=4)
simpleaxis(ax)
plt.xticks(rotation=30);
plt.xlabel('');
plt.ylabel('Distance (pixels)');
plt.ylim([0, 500]);
```

```python
distance_1_2_melt = pd.concat([distance_1_melt, distance_2_melt])
fig = plt.figure(figsize=(8,5), dpi=150)
ax = sns.boxplot(x='variable', y='value', data=distance_1_2_melt, fliersize=0)
# ax = sns.violinplot(x='variable', y='value', data=distance_1_2_melt, cut=0.25, inner='box', scale='count')
ax = sns.swarmplot(x='variable', y='value', data=distance_1_2_melt, color="0.4", size=4)
simpleaxis(ax)
plt.xticks(rotation=30);
plt.xlabel('');
plt.ylabel('Distance (pixels)');
plt.ylim([0, 500]);
```
